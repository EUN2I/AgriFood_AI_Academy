## 3. 교착어 vs. 고립어

<img src = "https://t1.daumcdn.net/cfile/tistory/9928F63D5B4974BB45" width = 500>

- 한국어의 말 단위 = 자립 형태소 + 의존 형태소로 구성 
- 자립 형태소 = 체언 (명사, 대명사, 수사), 수식언 (관형사, 부사), 감탄사
- 의존 형태소 = 접사, 어미, 조사, 어간 등
- 반면, 영어는 말 단위 = 단어이므로 띄어쓰기 단위로 토큰화 진행
- 한국어는 영어와 달리 띄어쓰기 단위의 토큰화가 아닌 형태소 단위의 토큰화가 중요

- 예) 제임스는 수지를 사랑한다
> - 자립 형태소 = 제임스, 수지, 사랑
> - 의존 형태소 = 는, 를, 한다

## 4. 표제어 추출 (Lemmatization)

- 표제어 (lemma): 기본 사전형 단어, 올림말이라고도 불림

- 영어에서는 시제, 단복수 등으로 인해 변하지 않는 원형

- (질문) '고춧가루'에서 ‘가루’는 붙여쓰는데 왜 '좁쌀 가루'에서 ‘가루’는 띄어쓰나요? 
- (답변) '가루’는 결합하는 단어들이 국어사전에 표제어 혹은 예시로 올라있으면 붙여 쓰고, 그렇지 않으면 띄어씁니다. 고춧가루는 예시로 올라있기에 붙여쓰고, 좁쌀 가루에서 ‘가루’는 예시로 올라있지 않기 때문에 띄어씁니다. 비슷한 예로는 담배 가루, 분필 가루 등이 있습니다.

## 5. 말뭉치와 말뭉치 주석

#### 1) 말뭉치 (corpus)
- 말뭉치 또는 코퍼스(영어: corpus, 복수형: corpora)는 자연 언어 (natural language)의 분석을 위해 특정 목적을 가지고 언어의 표본을 추출한 집합
- 예시: 각 단어의 품사 표기(동사, 명사, 형용사 등)에 대한 정보가 말뭉치에 '태그'의 형태로 추가되는 '품사 표기(part-of-speech tagging, POS-tagging)' 말뭉치

#### 2) 말뭉치 주석
- 말뭉치에 특별한 표시를 하는(tagging) 작업
- 품사, 감정, 성별, 실제 의도 등 말뭉치 개발 목적에 맞는 다양한 정보를 표시 (annotation)

#### 3) 국립국어원 말뭉치 예시와 통계

## 공백 제거

```python
# replace()

a = "우리 마트는 바나나를 판매합니다"

a.replace(" ", "")
```

```python
# strip()

a = "      우리 마트는 바나나를 판매합니다       " #문장 양쪽 공백

a.strip()
```

```python
# lstrip()

a = "      우리 마트는 바나나를 판매합니다       " #문장 left 공백

a.lstrip()
```

```python
# rstrip()

a = "      우리 마트는 바나나를 판매합니다       " #문장 right 공백

a.rstrip()
```

```python
# 정규표현식 사용

import re 

a = "      우리 마트는 바나나를 판매합니다       " 

print('문장의 모든 공백을 제거한 경우:', re.sub("\s", "", a))         #문장 모든 공백
print("---")
print('문장의 양쪽 공백을 제거한 경우:', re.sub("^\s+|\s+$", "", a))  #문장 양쪽 공백
print("---")
print('문장의 오른쪽 공백을 제거한 경우:', re.sub("\s+$", "", a))      #문장 right 공백
print("---")
print('문장의 왼쪽 공백을 제거한 경우:', re.sub("^\s+", "", a))       #문장 left 공백
```

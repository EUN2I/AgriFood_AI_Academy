```python
print(type(response.text))

contents = response.text

# !pip install xmltodict

import xmltodict
xmlDict = xmltodict.parse(contents)

# Print the dictionary
xmlDict['response']['body']['items']['item']

# 데이터프레임으로 변환
df = pd.DataFrame.from_dict(xmlDict['response']['body']['items']['item'], orient='columns')

df
```

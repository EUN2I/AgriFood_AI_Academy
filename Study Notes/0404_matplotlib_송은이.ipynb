```python
# plot() 함수 
# 기본: 선 그래프

x = np.array([0, 6])
y = np.array([0, 6])

plt.plot(x, y)
plt.show() # plt.show 코드 습관 만들기

# plt.plot(x, y)에 담긴 정보

print(plt.plot(x, y))

#Line2D
```

```python
# arange() 사용
x1 = np.arange(7)
y1 = np.arange(7)

plt.figure(figsize=(12,12)) #개별 그래프 크기 설정 #plot() 전에 호출해야

plt.plot(x1, y1, marker ='x', linestyle = '--')
plt.show()
```

```python
# 점 그래프

x2 = np.array([0, 100])
y2 = np.array([0, 100])

plt.plot(x2, y2, 'o') #ring 
plt.show()
```

```python
# 꺽은 선 그래프

x3 = np.array([4, 10, 40, 45, 54])
y3 = np.array([0, 23, 34, 78, 90])

plt.plot(x3, y3)
plt.show()
```

```python
# 꺽은 선 그래프 - y값만 존재 - x값 [0,1,2,3] 등으로 자동 설정

y4 = np.array([0, 23, 34, 78, 90])

plt.plot(y4)
plt.show()
```

```python
# 한글 깨짐 chunk 

x10 = np.array(['1월', '2월', '3월', '4월', '5월'])
y10 = np.array([0, 23, 34, 78, 90])

plt.figure(figsize=(9,6))

plt.plot(x10, y10, label = "월 강수량") #레전드
plt.legend()

plt.title("2022년 상반기 강수량")
plt.xlabel("x축의 이름입니다")
plt.ylabel("y축의 이름입니다")

plt.show()
```

```python
#폰트 설정

x10 = np.array(['1월', '2월', '3월', '4월', '5월'])
y10 = np.array([0, 23, 34, 78, 90])

# 그래프 디자인 설정 미리 해두기

fd1 = {'color':'m','size':35, 'weight': 'bold'}
fd2 = {'color':'blue','size':25}
fd3 = {'color':'red','size':25}

plt.figure(figsize=(9,6)) #plot() 전에 호출

plt.plot(x10, y10, label = "월 강수량") #레전드
plt.legend()

plt.title("2022년 상반기 강수량", fontdict = fd1, loc = 'right')
plt.text(1, 40, "지속 증가 추세", fontdict = fd2, rotation=30)  # -> 그래프 옆에 설명

plt.xlabel("x축의 이름입니다", fontdict = fd3)
plt.ylabel("y축의 이름입니다")

# plt.grid() # 격자그리기

# plt.grid(axis = 'x') #x축 격자만 표시
# plot.grid (axis = 'y') #y축 격자 표시

plt.grid(axis = 'x',      #격자 축
         color = 'b',     #격자 색깔
         linestyle = ':', #격자 스타일
         linewidth = 4.5) #격자 두께

plt.show()
```

````python
# continue 와 조건문 ("나는 스킵하고 계속 진행하세요")

apple = 0
while apple < 6:
  apple += 1
  if apple == 3:
    continue
  print(apple)
````

````python
# 딕셔너리 키 프린트
for i in zoo:
  print(i)

print("------------")

# 딕셔너리 값 프린트
for i in zoo.values():
  print(i)

print("------------")

# 딕셔너리 키, 값 같이 출력 
for i in zoo:
  print(i, zoo[i])

print("------------")

for i in zoo.items():
  print(i)

print("------------")

# enumerate() # 인덱스와 해당값을 일일이 나열


for i,v in enumerate(zoo.items()):
  print(i,v)
````

````python
apple = {'맥북':2021,'아이패드':2022,'아이맥':2023}

apple_2022 = {k:v for k,v in apple.items() if v <=2022}
apple_2022
````

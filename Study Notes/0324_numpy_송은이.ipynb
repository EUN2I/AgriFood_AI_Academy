```python
# 요소별로 계산한다 = 브로드케스팅

# 리스트 -> 요소별 계산 불가

print([10,20,30]+[1,2,3])

# np.array -> 요소별 계산 가능

print(np.array([10,20,30])+np.array([1,2,3]))
```

```python
b23=np.array([[4,5,6],[100,110,120]])
c23=b23.copy()
c23[1,2]=400
b23
```

```python
c23=b23
c23[1,2]=400
b23

# 이런식으로 복사하면 b23(원본)까지 변경됨 a=b, a=b.view() 같은 의미임
# .copy() 방식으로 복사해야 원본은 변경 안됨
```

```python
b23=np.array([[4,5,6],[100,110,120]])
c23=b23.view()
c23[1,2]=400
b23
```

```python
# axis 0 은 칼럼 기준 / axis 1 은 행 기준
```

```python
b23=np.array([[4,5,6],[100,110,120]])
print(b23.sum()) # 요소 전체를 한꺼번에 더하기

print(b23.sum(axis=0)) # 칼럼별로 더하기

print(b23.sum(axis=1)) # 행별로 더하기
```
